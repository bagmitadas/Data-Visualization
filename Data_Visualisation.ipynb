Question 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate correlated data
cov_matrix = np.array([[1, 0.8, 0.5],
                     [0.8, 1, 0.6],
                     [0.5, 0.6, 1]])
data = np.random.multivariate_normal([0, 0, 0], cov_matrix, size=100)
df = pd.DataFrame(data, columns=['Feature1', 'Feature2', 'Feature3'])

# Calculate correlation matrix
corr_matrix = df.corr()

# Visualize correlation matrix using heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

Question 3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
np.random.seed(20)
data = {
 'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
 'Day': np.random.choice(range(1, 31), 100),
 'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

```python
import numpy as np
import pandas as pd
import plotly.express as px

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivot the data for heatmap
df_pivot = df.pivot_table(values='Sales', index='Month', columns='Day')

# Create the heatmap
fig = px.imshow(df_pivot, text_auto=True, aspect="auto")
fig.update_layout(
    title="Sales Heatmap",
    xaxis_title="Day",
    yaxis_title="Month"
)
fig.show()
```

Question 4.  Using the given x and y data, generate a 3D surface plot to visualize the function. z = sin(sqrt(x**2+y**2)) Using the given x and y data, generate a 3D surface plot to visualize the function

```python
import numpy as np
import plotly.graph_objects as go

# Generate x and y data
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)

# Calculate z values
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create the surface plot
fig = go.Figure(data=[go.Surface(z=Z, x=x, y=y)])
fig.update_layout(title='Surface Plot of sin(sqrt(x^2 + y^2))')
```

Question 1. .Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Set output to notebook
output_notebook()

#Generate x and y values
x = np.linspace(0, 10, 1000)
y = np.sin(x)

# Create the plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

# Add a line renderer
p.line(x, y, line_width=2)

# Show the plot
show(p)
```

Question 2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
import numpy as np

# Generate random data
N = 50
x = np.random.random(size=N)
y = np.random.random(size=N)
sizes = np.random.randint(10, 50, size=N)
colors = ["#%02x%02x%02x" % tuple(np.random.randint(0, 256, 3)) for _ in range(N)]

# Create ColumnDataSource
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create the figure
p = figure(title="Scatter Plot using randomly generated values")

# Add scatter plot
p.scatter(x='x', y='y', size='sizes', color='colors', alpha=0.8, source=source)

# Show the plot
show(p)
```

Question 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

# Sample data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create the figure
p = figure(x_range=fruits, title="Fruit Counts")

# Add a bar renderer
p.vbar(x='fruits', top='counts', width=0.5, source=source)

# Show the plot
show(p)
```

Quetion 4. Create a Bokeh histogram to visualize the distribution of the given data.
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
import numpy as np

# Generate data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create ColumnDataSource
source = ColumnDataSource(data=dict(top=hist, bottom=np.zeros_like(hist), left=edges[:-1], right=edges[1:]))

# Create the figure
p = figure(title="Histogram")

# Add a bar renderer
p.quad(source=source)

# Show the plot
show(p)
```

Question 5. Create a Bokeh heatmap using the provided dataset.
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
import numpy as np

# Generate data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the data for ColumnDataSource
data = {'x': xx.flatten(), 'y': yy.flatten(), 'z': data_heatmap.flatten()}
source = ColumnDataSource(data)

# Create the figure
p = figure(title="Heatmap")

# Add the heatmap
p.rect(x='x', y='y', width=0.1, height=0.1, source=source, fill_color='z', fill_alpha=0.8)

# Show the plot
show(p)
```
